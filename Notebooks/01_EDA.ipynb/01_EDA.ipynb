# Movie Insights — Análisis Exploratorio de Datos (EDA)

Este notebook forma parte del proyecto **Movie Insights**, cuyo objetivo es 
explorar y entender el comportamiento de las películas registradas en 
TMDB (The Movie Database) entre 2020 y 2025.

El análisis exploratorio nos permite conocer la estructura de los datos, 
detectar problemas de calidad (nulos, outliers, tipos incorrectos) y 
obtener una primera visión estadística antes de construir visualizaciones 
o modelos más complejos.

## 1. Carga de datos

Cargamos el archivo CSV ya procesado (`tmdb_limpio.csv`), que contiene 
información de películas como título, idioma, fecha de lanzamiento y 
métricas de valoración de los usuarios.

```python
import pandas as pd

df = pd.read_csv(r'C:\Movie_Insights\Data\processed\tmdb_limpio.csv')

print(f"Dimensiones del dataset: {df.shape[0]} filas x {df.shape[1]} columnas")
df.head()
```

## 2. Resumen Descriptivo

El resumen descriptivo nos da informacion estadística de las variables 
numéricas del dataset. Calculamos las siguientes medidas para cada variable:

| Medida | Descripción |
|--------|-------------|
| `count` | Número de valores no nulos |
| `mean` | Promedio aritmético |
| `std` | Desviación estándar (qué tan dispersos están los datos) |
| `min` / `max` | Valor mínimo y máximo |
| `Q1` / `Q3` | Percentil 25 y 75 (rango intercuartílico) |
| `median` | Valor central de la distribución |

En este dataset las variables numéricas son `vote_average` (calificación 
promedio de 0 a 10) y `vote_count` (cantidad de votos recibidos).

```python
numericas = df.select_dtypes(include='number')

resumen = numericas.agg([
    'count', 'mean', 'std', 'min',
    lambda x: x.quantile(0.25),
    'median',
    lambda x: x.quantile(0.75),
    'max'
]).round(2)

resumen.index = ['count', 'mean', 'std', 'min', 'Q1', 'median', 'Q3', 'max']
resumen
```

## 3. Interpretación de resultados en cuanto al resumen descriptivo

### vote_average (Calificación promedio)

- La calificación **promedio es 4.55**, pero la **mediana es 5.71**, lo que 
  indica que la distribución está dividida hacia la izquierda: hay muchas 
  películas con calificación 0 que jalan el promedio hacia abajo.
- El **Q1 es 0.0**, lo que significa que al menos el 25% de las películas 
  no tienen ninguna calificación registrada todavía.
- El rango va de **0 a 10**, cubriendo toda la escala posible.
- La desviación estándar de **2.97** indica alta variabilidad entre películas.

## 4. Matriz de Correlación

La matriz de correlación mide qué tan relacionadas están dos variables 
numéricas entre sí. El coeficiente va de **-1 a 1**:

| Valor | Interpretación |
|-------|---------------|
| Cercano a 1 | Correlación positiva fuerte |
| Cercano a -1 | Correlación negativa fuerte |
| Cercano a 0 | Sin correlación |

Analizamos si existe relación entre la **calificación promedio** 
(`vote_average`) y la **cantidad de votos** (`vote_count`).

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlacion = df[['vote_average', 'vote_count']].corr()

print(correlacion)

plt.figure(figsize=(6, 4))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt='.2f', 
            linewidths=0.5, vmin=-1, vmax=1)
plt.title('Matriz de Correlación')
plt.tight_layout()
plt.show()
```

### Interpretación

La correlación entre `vote_average` y `vote_count` es **0.20**, lo que indica 
una correlación positiva muy débil. Esto significa que existe una ligera 
tendencia a que las películas con más votos tengan calificaciones más altas, 
pero la relación es tan débil que prácticamente no hay conexión entre ambas variables.

Es decir que una película tenga muchos votos **no garantiza** 
que sea bien calificada, ni que una película bien calificada vaya a recibir 
muchos votos. Son variables casi independientes entre sí.

Por otro lado la diagonal muestra **1.00** porque cualquier variable tiene 
correlación perfecta consigo misma.


## 5. Análisis de Outliers

Un outlier es un valor atípico que se aleja significativamente del resto. 
Usamos el método **IQR (Rango Intercuartílico)** ya que se considera outlier 
cualquier valor por debajo de `Q1 - 1.5*IQR` o por encima de `Q3 + 1.5*IQR`.

Detectar outliers es importante porque pueden distorsionar promedios 
y modelos posteriores.

```python
def detectar_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR
    outliers = df[(df[columna] < limite_inf) | (df[columna] > limite_sup)]
    print(f"--- {columna} ---")
    print(f"  Límite inferior: {limite_inf:.2f}")
    print(f"  Límite superior: {limite_sup:.2f}")
    print(f"  Outliers encontrados: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")
    return outliers

outliers_avg   = detectar_outliers(df, 'vote_average')
outliers_count = detectar_outliers(df, 'vote_count')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].boxplot(df['vote_average'].dropna(), patch_artist=True,
                boxprops=dict(facecolor='#e74c3c', alpha=0.7),
                medianprops=dict(color='#7b0000', linewidth=2),
                whiskerprops=dict(color='#c0392b'),
                capprops=dict(color='#c0392b'),
                flierprops=dict(markerfacecolor='#e74c3c', marker='o'))
axes[0].set_title('Outliers — vote_average')
axes[0].set_ylabel('Calificación')

axes[1].boxplot(df['vote_count'].dropna(), patch_artist=True,
                boxprops=dict(facecolor='#f1948a', alpha=0.7),
                medianprops=dict(color='#7b0000', linewidth=2),
                whiskerprops=dict(color='#c0392b'),
                capprops=dict(color='#c0392b'),
                flierprops=dict(markerfacecolor='#f1948a', marker='o'))
axes[1].set_title('Outliers — vote_count')
axes[1].set_ylabel('Cantidad de votos')

plt.tight_layout()
plt.show()
```

### Interpretación

- **vote_average**: Los outliers son películas con calificación 0, que 
  corresponden a estrenos sin votos aún, no a películas  
  malas. 

- **vote_count**: Hay outliers extremos hacia arriba, películas con 
  muchos votos. Ademas el boxplot muestra una caja muy comprimida 
  cerca del 0, lo que confirma que la gran mayoría de películas tienen 
  muy pocos votos y unas pocas tienen toda la actividad.

### Filtrado de outliers para análisis de calidad

Como se observó, muchas películas tienen `vote_average = 0` y `vote_count = 0`,
lo que no indica que sean malas, sino que simplemente **aún no tienen votos**.
Incluirlas en análisis de calidad distorsionaría los resultados.

Por eso creamos un DataFrame filtrado que excluye estas películas, 
conservando solo las que tienen al menos 10 votos.

```python
# Filtramos películas sin votos
df_calidad = df[df['vote_count'] >= 10].copy()

print(f"Dataset original:  {len(df)} películas")
print(f"Dataset filtrado:  {len(df_calidad)} películas")
print(f"Películas removidas: {len(df) - len(df_calidad)} ({(len(df) - len(df_calidad))/len(df)*100:.1f}%)")
print()
print("Resumen vote_average después del filtro:")
print(df_calidad['vote_average'].describe().round(2))
```

### Interpretación del filtrado

Al aplicar el filtro de `vote_count >= 10`, el dataset se redujo de 
**9,999 a 4,499 películas**, eliminando el **55%** de los registros. 
Esto confirma que más de la mitad del dataset corresponde a estrenos 
recientes o producciones sin actividad de votación significativa.

El impacto en la calificación promedio es notable:

| Medida | Sin filtro | Con filtro |
|--------|-----------|------------|
| Media | 4.55 | 6.30 |
| Desviación estándar | 2.97 | 0.94 |
| Mínimo | 0.00 | 1.80 |
| Mediana | 5.71 | 6.38 |

La media **subió casi 2 puntos** y la desviación estándar bajó de 2.97 
a 0.94, lo que indica que los datos filtrados son mucho más consistentes 
y representativos de la opinión real del público.


## 6. Análisis por Año de Lanzamiento

Extraemos el año de `release_date` para analizar cómo ha evolucionado 
las películas y sus calificaciones entre 2020 y 2025.

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['anio'] = df['release_date'].dt.year

peliculas_por_anio = df.groupby('anio').agg(
    total_peliculas=('title', 'count'),
    calificacion_promedio=('vote_average', 'mean'),
    votos_promedio=('vote_count', 'mean')
).round(2)

print(peliculas_por_anio)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

peliculas_por_anio['total_peliculas'].plot(kind='bar', ax=axes[0], 
                                            color='#c0392b', alpha=0.8)
axes[0].set_title('Películas por año')
axes[0].set_xlabel('Año')
axes[0].set_ylabel('Total')

peliculas_por_anio['calificacion_promedio'].plot(kind='line', ax=axes[1], 
                                                  marker='o', color='salmon')
axes[1].set_title('Calificación promedio por año')
axes[1].set_xlabel('Año')
axes[1].set_ylabel('Calificación')

plt.tight_layout()
plt.show()
```

### Interpretación

Este análisis permite ver si la cantidad de películas producidas creció o 
disminuyó en el período, y si las calificaciones han mejorado o empeorado 
con el tiempo. Un dato que llama la atencion es que en en 2022 fue el año con mas peliculas producidas y con mejor calificacion promedio, por otro lado tenemos que en los años que las personas mas votaron fue del 2020 al 2022 una hipotesis de esto es que en esos años nos encontrabamos en pandemia y las personas estaban en sus casas lo que pudo llegar a ayudar e incrementar la interaccion de las personas con las peliculas.
